Bài 1: Sinh mẫu ngẫu nhiên bằng hàm arange trong khoảng [-15, 27] với khoảng cách là 0.7 và in kết quả ra màn hình

```python
# Sử dụng thư viện Numpy
low = -15
high = 27
dis = math.sqrt(2)
test1 = np.arange(low, high, dis)
test1.sort()
data1 = pd.DataFrame(test1)
data1.columns = ['Arange']
data1
```

Bài 5: Sinh mẫu ngẫu nhiên theo phân phối chuẩn với mean = 12.7, std = 2.6 và size = 30. In kết quả ra màn hình

```python
# Sử dụng thư việc scipy.stats để sinh mẫu ngẫu nhiên
mean5 = 12.7
std5 = 2.6
size5 = 30
test5 = stats.norm.rvs(mean5, std5, size5)
test5.sort()
data5 = pd.DataFrame(test5, columns = ['Scipy.stats'])
data5
```

```python
#Sinh ngẫu nhiên dữ liệu theo phân phối Poisson
test6 = stats.poisson.rvs(mean5, std5, size5)
test6.sort()
data6 = pd.DataFrame(test6, columns = ['Scipy.stats Poisson'])
data6
```
